```python
# Data Augmentation

# Its all about manipulating the image before giving input to the NN
# Manipulation includes
# - Scaling
# - Shrinking
# - Rotation
# - Shear

#Goal: To make the data compatible for the NN
```

```python
import tensorflow as tf
```

```python
#Load the Image
img = tf.keras.preprocessing.image.load_img("/content/drive/MyDrive/DA Day1/Day2/cat.103.jpg", target_size=(256,256))
```

```python
#Visualize
import matplotlib.pyplot as plt
%matplotlib inline
plt.imshow(img)
```

```python
# Convert the image into numpy array
imgArray = tf.keras.preprocessing.image.img_to_array(img)
```

```python
imgArray.shape
```

```python
#Make the image compatible for NN

import numpy as np
compatibleImage = np.expand_dims(imgArray, axis=0)
compatibleImage.shape
```

```python
#Data Augmentation in Action

train_generator = tf.keras.preprocessing.image.ImageDataGenerator(rotation_range=90,
                                                                  zoom_range=0.2,
                                                                  horizontal_flip=True,
                                                                  vertical_flip=True)


#Generate Image

imgGenerate = train_generator.flow(compatibleImage,
                                   batch_size=1,
                                   save_to_dir='AugmentedImage',
                                   save_prefix='cats_',
                                   save_format='jpg')
```

```python
#Create 10 Image
counter = 1
for img in imgGenerate:
  counter+=1
  if counter == 10:
    break
```
