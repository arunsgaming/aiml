```python
import tensorflow as tf
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tf.__version__
```

```python
#Load the model
model = tf.keras.applications.ResNet50(weights="imagenet")
```

```python
#Load an image
imageToPredict = tf.keras.preprocessing.image.load_img("elephant.jpg", target_size=(224,224))
plt.imshow(imageToPredict)
```

```python
#Convert Image to a NP array
myImageToNPArray = tf.keras.preprocessing.image.img_to_array(imageToPredict)
myImageToNPArray.shape
```

```python
#Make the data compatible for teh model
myCompatibleImage = np.expand_dims(myImageToNPArray, axis=0)
myCompatibleImage.shape
```

```python
#Make a Prediction
prediction = model.predict(myCompatibleImage)
tf.keras.applications.resnet50.decode_predictions(prediction)
```

```python
#Perform teh same exercise with the following architecture
#
# 1. VGG16
# 2. Xception
# 3. Inception(anyversion)
# 4. NasNet
# 5. MobileNet
```

```python
# 1. VGG16
modelVGG = tf.keras.applications.VGG16(weights="imagenet")
```

```python
#Make a Prediction
imageToPredict = tf.keras.preprocessing.image.load_img("elephant.jpg", target_size=(224,224))
myImageToNPArray = tf.keras.preprocessing.image.img_to_array(imageToPredict)
myCompatibleImage = np.expand_dims(myImageToNPArray, axis=0)
prediction = modelVGG.predict(myCompatibleImage)
tf.keras.applications.vgg16.decode_predictions(prediction)
```

```python
# 2. Xception
modelXception = tf.keras.applications.Xception(weights="imagenet")
```

```python
#Make a Prediction
imageToPredict = tf.keras.preprocessing.image.load_img("elephant.jpg", target_size=(299,299))
myImageToNPArray = tf.keras.preprocessing.image.img_to_array(imageToPredict)
myCompatibleImage = np.expand_dims(myImageToNPArray, axis=0)

prediction = modelXception.predict(myCompatibleImage)
tf.keras.applications.xception.decode_predictions(prediction)
```

```python
# 3. Inception(anyversion)

modelInception = tf.keras.applications.InceptionV3(weights="imagenet")

#Make a Prediction
imageToPredict = tf.keras.preprocessing.image.load_img("elephant.jpg", target_size=(299,299))
myImageToNPArray = tf.keras.preprocessing.image.img_to_array(imageToPredict)
myCompatibleImage = np.expand_dims(myImageToNPArray, axis=0)

prediction = modelInception.predict(myCompatibleImage)
tf.keras.applications.inception_v3.decode_predictions(prediction)
```

```python
# 4. NasNet

modelNasnet = tf.keras.applications.nasnet.NASNetLarge(weights='imagenet')

#Make a Prediction
imageToPredict = tf.keras.preprocessing.image.load_img("elephant.jpg", target_size=(331,331))
myImageToNPArray = tf.keras.preprocessing.image.img_to_array(imageToPredict)
myCompatibleImage = np.expand_dims(myImageToNPArray, axis=0)

prediction = modelNasnet.predict(myCompatibleImage)
tf.keras.applications.nasnet.decode_predictions(prediction)
```

```python
# 5. MobileNet

modelMobilenet = tf.keras.applications.mobilenet.MobileNet(weights='imagenet')

#Make a Prediction
imageToPredict = tf.keras.preprocessing.image.load_img("elephant.jpg", target_size=(224,224))
myImageToNPArray = tf.keras.preprocessing.image.img_to_array(imageToPredict)
myCompatibleImage = np.expand_dims(myImageToNPArray, axis=0)

prediction = modelMobilenet.predict(myCompatibleImage)
tf.keras.applications.nasnet.decode_predictions(prediction)
```
